```python
import numpy
import matplotlib.pyplot
```

```python
qdata = numpy.loadtxt(fname = '080167500_Guadalupe_SpringBranch_DailyMeanQ.csv', delimiter = ',')
```

```python
print(type(qdata))
```

```python
maxval, minval, stdval, qrange = numpy.max(qdata), numpy.min(qdata), numpy.std(qdata), numpy.ptp(qdata, axis=0)
qrange2 = (maxval-minval)
```

```python
print('maximum of Q', maxval)
print('minimum of Q:', minval)
print('standard deveiation of Q:', stdval)
print('range of Q days?:', qrange)
print(numpy.mean(qdata, axis = 0).shape)
print('range of Q days?:', qrange2)
```

```python
Qfig = matplotlib.pyplot.figure(figsize=(10.0, 3.0))

axes1 = Qfig.add_subplot(1, 5, 1)
axes2 = Qfig.add_subplot(1, 5, 2)
axes3 = Qfig.add_subplot(1, 5, 3)
axes4 = Qfig.add_subplot(1, 5, 4)
axes5 = Qfig.add_subplot(1, 5, 5)

axes1.set_ylabel('average')
axes1.plot(numpy.mean(qdata, axis=0))

axes2.set_ylabel('max')
axes2.plot(numpy.max(qdata, axis=0))

axes3.set_ylabel('min')
axes3.plot(numpy.min(qdata, axis=0))

axes4.set_ylabel('range')
axes4.plot(numpy.max(qdata, axis=0) - numpy.min(qdata, axis=0))

axes5.set_ylabel('range 2')
axes5.plot(numpy.ptp(qdata, axis=0))

Qfig.tight_layout()
matplotlib.pyplot.show()
```
